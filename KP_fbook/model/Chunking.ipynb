```python
import nltk
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.tokenize import RegexpTokenizer
from clean_text import clean
import pandas as pd
tf_idf = pd.read_csv('results/tf_idf_evaluation.csv')
text = tf_idf['threads'][14]
sentences = sent_tokenize(text)

clean_sentences = [clean(sentence) for sentence in sentences]


for i, sent in enumerate(clean_sentences, 1):
    print('sentence: {} {}\n'.format(i, sent))

```

```python
sent = clean_sentences[4]
#Noun Phrase chunking
#sent = "The little mouse ate the fresh cheese"
sent_tokens = nltk.pos_tag(word_tokenize(sent))
grammar_np = r"NP:{<DT>?<JJ>*<NN>}"
chunk_parser = nltk.RegexpParser(grammar_np)
chunk_result = chunk_parser.parse(sent_tokens)
chunk_result
```

```python
#Verb Phrase chunking
sent = clean_sentences[4]


#sent2 = "She is walking quickly to the mall"
#sent2_tokens = nltk.pos_tag(word_tokenize(sent2))
grammar_vp = r"VP:{<PRP>?<VB>*<RB|RBR>?}"
chunk2_parser = nltk.RegexpParser(grammar_vp)
chunk2_result = chunk2_parser.parse(sent_tokens)
chunk2_result
```

```python
#adjective phrase
#Playing around with this https://grammar.yourdictionary.com/parts-of-speech/adjectives/adjective-phrase.html

#sent3 = "The dog that seems very angry barks all day long."
#sent3_tokens = nltk.pos_tag(word_tokenize(sent3))

sent = clean_sentences[4]


grammar_ap = r"AP:{<RB>+<JJ>}"
chunk3_parser = nltk.RegexpParser(grammar_ap)
chunk3_results = chunk3_parser.parse(sent_tokens)
chunk3_results
```
